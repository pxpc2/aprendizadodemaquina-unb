```python
path = untar_data(URLs.ML_100k)
Path.BASE_PATH = path

path.ls().sorted()
```

```python
ratings = pd.read_csv(path/'u.data', delimiter='\t', header=None,
                      names=['user', 'movie', 'rating', 'timestamp'])
ratings.head()
```

```python
movies = pd.read_csv(path/'u.item', delimiter='|',
                     encoding='latin-1', usecols=(0,1),
                     names=('movie','title'), header=None)

movies.head()
```

```python
ratings = ratings.merge(movies)

ratings.head()
```

Por padrão, ao criar um Dls a partir de uma tabela do pandas, a primeira coluna é o user, e segunda coluna seria o item, e terceira a ratings. Vamos alterar o valor de item_name para usar title invés do movie_id:

```python
dls = CollabDataLoaders.from_df(ratings, item_name='title', bs=64)
dls.show_batch()
```
